### Sample Request

Let's submit a sample request to the Hugging Face Inference endpoint!

```python
import requests

max_new_tokens = 256
top_p = 0.9
temperature = 0.1
model_api_gateway = "https://meqlao8jlyg1d5w2.us-east-1.aws.endpoints.huggingface.cloud/"

prompt = "Hello! How are you?"

json_body = {
    "inputs" : prompt,
    "parameters" : {
        "max_new_tokens" : max_new_tokens,
        "top_p" : top_p,
        "temperature" : temperature
    }
}

headers = {
  "Authorization": f"Bearer {os.environ['HF_TOKEN']}",
  "Content-Type": "application/json"
}

response = requests.post(model_api_gateway, json=json_body, headers=headers)
print(response.json())
```

Let's SLAM THAT ENDPOINT!

```python
import nest_asyncio
nest_asyncio.apply()
```

```python
import asyncio
import aiohttp

async def call_endpoint(model_api_gateway, json_body, headers):
    async with aiohttp.ClientSession() as session:
        async with session.post(model_api_gateway, json=json_body, headers=headers) as response:
            return await response.json()

async def main():
    tasks = []
    for _ in range(100):
        task = asyncio.create_task(call_endpoint(model_api_gateway, json_body, headers))
        tasks.append(task)

    results = await asyncio.gather(*tasks)
    for result in results:
        print(result)

asyncio.run(main())
```
